```python
marks = [78, 85, 90, 66, 88]
print(f"Maximum marks: {max(marks)}")
print(f"Minimum marks: {min(marks)}")
print(f"Total marks: {sum(marks)}")
print(f"Number of students: {len(marks)}")
print(f"Average marks: {sum(marks) / len(marks)}")
marks.sort()
print(f"Marks sorted in ascending order: {marks}")
```

#Grocery List

#Aim
The goal of this code is to modify a predefined list. It demonstrates how to start with a basic grocery list and then update it by adding new items and removing specific ones to keep the list current.

#Flowchart
Step 1: Create a list with 5 items (Bread, Milk, Eggs, Cereal, Apples).

Step 2: Add "Bananas" and "Flour" to the end of that list.

Step 3: Show the updated list on the screen.

Step 4: Delete "Apples" and "Bread" from the list.

Step 5: Show the final version of the list.

```python
Grocery=['Bread','Milk','Eggs','Cereal','Apples']
Grocery.append('Bananas')
Grocery.append('Flour')
print(Grocery)
Grocery.remove('Apples')
Grocery.remove('Bread')
print(Grocery)
```

#Conclusion
This script shows how Python handles dynamic data. It proves that lists are not permanent; you can easily expand them using .append() or shrink them using .remove(). By the end of the script, the computer has successfully updated the "Grocery" variable to reflect the changes.
